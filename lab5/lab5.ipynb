1. Створити набір даних за допомогою DatasetGenerator



Генерування набору даних за допомогою make_regression, де проводиться регресійний аналіз.
Для отриманих наборів даних прововодимо встановлення моделювання.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Генерація набору даних
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Створення DataFrame
df = pd.DataFrame(X, columns=['X'])
df['y'] = y

# Виведення перших кількох рядків набору даних
print(df.head())

# Створення і навчання моделі лінійної регресії
model = LinearRegression()
model.fit(df[['X']], df['y'])
y_pred = model.predict(df[['X']])

# Оцінка якості моделі
mse = mean_squared_error(df['y'], y_pred)
mae = mean_absolute_error(df['y'], y_pred)
r2 = r2_score(df['y'], y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {r2}')

# Візуалізація результатів
plt.scatter(df['X'], df['y'], color='blue', label='Actual')
plt.plot(df['X'], y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія (make_regression)')
plt.legend()
plt.show()
```

Генерування набору даних для регресії, використовуючи стандартні можливості Python 

У цьому завданні ми згенеруємо дані для лінійної регресії вручну за допомогою numpy.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Генерація набору даних для регресії
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) * 0.5  # Лінійна залежність з шумом

# Створення DataFrame
df = pd.DataFrame(X, columns=['X'])
df['y'] = y

# Виведення перших кількох рядків набору даних
print(df.head())

# Створення і навчання моделі лінійної регресії
model = LinearRegression()
model.fit(df[['X']], df['y'])
y_pred = model.predict(df[['X']])

# Оцінка якості моделі
mse = mean_squared_error(df['y'], y_pred)
mae = mean_absolute_error(df['y'], y_pred)
r2 = r2_score(df['y'], y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R^2 Score: {r2}')

# Візуалізація результатів
plt.scatter(df['X'], df['y'], color='blue', label='Actual')
plt.plot(df['X'], y_pred, color='red', label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія (генерація за допомогою numpy)')
plt.legend()
plt.show()


```
