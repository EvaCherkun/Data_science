Завдання 1:


* Токенізація: Текст розбивається на слова або речення.
* Лемматизація і стемінг:
* Лемматизація – це коли слово приводять до його базової форми (наприклад, “бігаю” → “біг”).
* Стемінг – це обрізання слова до основи (наприклад, “бігаю” → “біг”).
*  BOW (Bag of Words) – просто рахуємо, скільки разів кожне слово зустрічається.
*  TF-IDF – враховує частоту слова і наскільки воно важливе в тексті.
*  Word Embeddings (Word2Vec, GloVe) – це коли слова перетворюють у вектори з урахуванням контексту.
*  Класифікація тексту: Наприклад, розподіл текстів за категоріями або визначення тональності.
*  Розпізнавання сутностей (NER): Визначення у тексті імен, назв компаній, дат тощо.
*  Наївний баєсовий класифікатор – підходить для простих задач, наприклад, класифікації.
* Логістична регресія – ще один популярний метод для текстової класифікації.
*  LSTM – модель, яка добре працює з послідовностями, такими як текст.
*  Transformers – це сучасні моделі (наприклад, BERT, GPT). GPT використовується для створення текстів.
