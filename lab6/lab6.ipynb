Здійснити загальні операції з використанням TensorFlow (створити набір даних (розмір за власним рішенням) у вигляді тензорів, провести CRUD операції над даними (скаляр, вектор, матриця)).


CRUD операції з скаляром:

```python
import tensorflow as tf

# CREATE
scalar = tf.constant(10)

# READ
print(f"Scalar value: {scalar.numpy()}")

# UPDATE (створення нового)
updated_scalar = scalar + 5
print(f"Updated scalar value: {updated_scalar.numpy()}")

# DELETE виведення None або залишаємо без змін)
scalar = None
print(f"Scalar after deletion: {scalar}")
```

2. CRUD операції з вектором:

```python
# CREATE
vector = tf.constant([1, 2, 3, 4, 5])

# READ
print(f"Vector value: {vector.numpy()}")
 
# UPDATE (додавання 2 до кожного елемента)
updated_vector = vector + 2
print(f"Updated vector: {updated_vector.numpy()}")

# DELETE виведення None або залишаємо без змін)
vector = None
print(f"Vector after deletion: {vector}")
```

3. CRUD операції з матрицею:

```python
# CREATE
matrix = tf.constant([[1, 2], [3, 4], [5, 6]])

# READ
print(f"Matrix value:\n{matrix.numpy()}")

# UPDATE (додавання 10 до кожного елемента)
updated_matrix = matrix + 10
print(f"Updated matrix:\n{updated_matrix.numpy()}")

# DELETE ( виведення None або залишаємо без змін)
matrix = None
print(f"Matrix after deletion: {matrix}")
```

Генерація набору даних для регресії та класифікації:

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Генерація випадкових даних для регресії
X = np.linspace(-1, 1, 100)  # Вхідні дані (вектор)
Y = 3 * X + 2 + np.random.normal(0, 0.1, 100) 

# Перетворення вхідних і вихідних даних у тензори
X_tensor = tf.convert_to_tensor(X, dtype=tf.float32)
Y_tensor = tf.convert_to_tensor(Y, dtype=tf.float32)

# Виведення перших 5 значень тензорів
print("X_tensor (перші 5 значень):", X_tensor[:5].numpy())  
print("Y_tensor (перші 5 значень):", Y_tensor[:5].numpy())

# Візуалізація даних
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Data points')

# Налаштування заголовку та підписів
plt.title("Regression Data")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()

# Показ графіку
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Генерація даних для класифікації
X_class = np.random.randn(200, 2)  
y_class = np.zeros(200)  
X_class[:100] += 3  
y_class[100:] = 1  

# Створення тензорів для TensorFlow
X_class_tensor = tf.constant(X_class, dtype=tf.float32)
y_class_tensor = tf.constant(y_class, dtype=tf.float32)

# Виведемо перші 5 елементів
print("X_class:", X_class_tensor[:5])
print("y_class:", y_class_tensor[:5])

# Візуалізація даних для класифікації
plt.figure(figsize=(6, 6))
plt.scatter(X_class[:100, 0], X_class[:100, 1], color='green', label='Class 0')
plt.scatter(X_class[100:, 0], X_class[100:, 1], color='orange', label='Class 1')
plt.title("Classification Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

```
